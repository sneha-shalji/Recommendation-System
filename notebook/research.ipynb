```python
ratings = ratings[ratings['user-id'].isin(y)] #return us the rating data of the users that has read more than 200 books
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books,on='ISBN')
```

```python
ratings_with_books
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating': 'num_of_rating'},inplace=True)
```

```python
number_rating
```

```python
final_rating = ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating
```

```python
final_rating = final_rating[final_rating['num_of_rating']>= 50]
```

```python
final_rating
```

```python
final_rating.drop_duplicates(['user-id','title'],inplace=True)
```

```python
final_rating.isna().sum()
```

```python
# Now creating a Pivot Table
```

```python
book_pivot = final_rating.pivot_table(columns="user-id",index="title",values="rating")
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
# converting the pivot table into a CSR matrix
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm="brute")
```

```python
model.fit(book_sparse)
```
